```python
class Autism:
    def __init__(self, name, age, fluent_speech, communication_difficulty, autism_level, likes_video_games, school, receives_help, family_communication):
        if 3 <= age <= 10:
            self.name = name
            self.age = age
            self.fluent_speech = fluent_speech
            self.communication_difficulty = communication_difficulty
            self.autism_level = autism_level
            self.likes_video_games = likes_video_games
            self.school = school
            self.receives_help = receives_help
            self.family_communication = family_communication
            self.interests = []
        else:
            raise ValueError("Age should be between 3 and 10.")

    def speak_fluently(self):
        if self.fluent_speech:
            print(f"{self.name} speaks fluently.")
        else:
            print(f"{self.name} does not speak fluently.")

    def communicate_difficulty(self):
        if self.communication_difficulty:
            print(f"{self.name} has difficulty in communication.")
        else:
            print(f"{self.name} communicates effectively.")

    def get_age(self):
        print(f"{self.name} is {self.age} years old.")

    def get_autism_level(self):
        print(f"{self.name} has an autism level of {self.autism_level}.")

    def check_video_game_preference(self):
        if self.likes_video_games:
            print(f"{self.name} likes playing video games.")
        else:
            print(f"{self.name} does not enjoy playing video games.")

    def get_school(self):
        print(f"{self.name} attends {self.school}.")

    def get_help_status(self):
        if self.receives_help:
            print(f"{self.name} receives assistance.")
        else:
            print(f"{self.name} does not receive assistance.")

    def family_communication_status(self):
        if self.family_communication:
            print(f"The family communication of {self.name} is correct and comfortable.")
        else:
            print(f"The family communication of {self.name} is not correct and comfortable.")

    def add_interest(self, interest):
        self.interests.append(interest)

    def get_interests(self):
        if self.interests:
            print(f"{self.name}'s interests are: {', '.join(self.interests)}")
        else:
            print(f"{self.name} has no specified interests.")


# Ejemplo de uso
person = Autism("John", 5, True, False, "High", True, "ABC School", True, True)
person.speak_fluently()
person.communicate_difficulty()
person.get_age()
person.get_autism_level()
person.check_video_game_preference()
person.get_school()
person.get_help_status()
person.family_communication_status()

person.add_interest("Reading")
person.add_interest("Playing")
person.add_interest("Drawing")
person.get_interests()
```
